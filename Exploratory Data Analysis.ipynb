```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#read dataset 
df = pd.read_csv("data/creditcard.csv")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
for col in df.columns:
    plt.figure()
    plt.hist(df[col], bins=50)
    plt.xlabel(col)
    plt.ylabel('Frequency')
```

```python
print(df.corr())
```

```python
corr_mat = df.corr().to_numpy()
plt.figure(figsize = (20,20))
plt.matshow(corr_mat)
plt.colorbar()
```
